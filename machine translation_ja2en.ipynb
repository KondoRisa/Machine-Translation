```python
from pyknp import Juman
jumanpp = Juman()

def juman(outfile, infile):
    fout = open(outfile, "w")
    fin = open(infile, "r")
    for line in fin:
        fout.write(" ".join([mrph.midasi for mrph in jumanpp.analysis(line.strip()).mrph_list()]) + "\n")
    fin.close()
    fout.close()
```

```python
import zenhan

def sentencepiece(outfile, infile):
    fout = open(outfile, "w")
    fin = open(infile, "r")
    for line in fin:
        text = " ".join(sp.encode_as_pieces(line.strip()))
        fout.write(zenhan.h2z(text).replace('　', ' ') + "\n")
    fin.close()
    fout.close()
```

#共通の処理

```python
from torchtext.data.metrics import bleu_score

def bleu(fname_ref, fname_pred):    
    # 正解文
    fin = open(fname_ref, "r")
    refs = [[line.strip().split()] for line in fin]
    fin.close()
    # 出力文
    fin = open(fname_pred, "r")
    preds = [line.strip().split() for line in fin]
    fin.close()
    # BLEU
    score = bleu_score(preds, refs)
    print("BLEU = %.2f" % (score*100))
```
